## 8. Хитрый вопрос 

Правда ли, что при более сильном ветре человеку ощущается, что на улице холодне, чем на самом деле? Нарисуйте график, подтверждающий или опровергающий эту гипотезу. Обоснуйте. Возможно, вам придется создать новый признак.

Рассуждение:

В DataFrame у нас есть:

    temp — фактическая температура,

    atemp — температура, ощущаемая человеком (возможно, уже с учетом ветра и влажности),

    windspeed(ms) — скорость ветра в метрах в секунду.

Можно рассчитать разницу между temp и atemp, и посмотреть, как она зависит от скорости ветра:
delta=temp−atemp
delta=temp−atemp

Если при увеличении ветра delta растет, это означает, что ощущаемая температура ниже фактической при более сильном ветре, что подтверждает гипотезу

На графике линия тренда восходящая - что подтверждает гипотезу.

```python
df["delta"] = df["temp"] - df["atemp"]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="windspeed(ms)", y="delta", alpha=0.6)
sns.regplot(data=df, x="windspeed(ms)", y="delta", scatter=False, color='red')

plt.title("Разница между фактической и ощущаемой температурой в зависимости от скорости ветра")
plt.xlabel("Скорость ветра (м/с)")
plt.ylabel("Фактическая - Ощущаемая температура (°C)")
plt.grid(True)
plt.show()
```
